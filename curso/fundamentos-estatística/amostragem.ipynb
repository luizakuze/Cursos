# Amostragem Aleatória Simples
Seleção por sorteio, sem filtros.

````python
np.random.seed(1) # gera valor aleatório e sempre que rodar vai dar o mesmo número
````

- np.random.seed é usado para inicializar o gerador de números aleatórios do NumPy com uma semente específica.
- Serve para reproduzir resultados aleatórios em diferentes execuções do código.
- A semente deve ser definida antes de qualquer chamada a funções que geram números aleatórios.
- Se definida corretamente, produzirá a mesma sequência de números pseudoaleatórios em diferentes execuções.
- Se definida após a geração de números aleatórios, não terá efeito retroativo na sequência já gerada.
- Se, mesmo após definir a semente, os resultados continuarem diferentes, revise o código para garantir que todas as fontes de aleatoriedade sejam tratadas adequadamente.

# Amostragem Estratificada
Divisão em grupos da população e seleção aleatória de uma amostra de cada grupo.

````python
enem_campinas['RACA'].value_counts() # analise estratificada para raça
````

````python
branca = enem_campinas.loc[enem_campinas.RACA == 'branca'].sample(frac=0.15)
branca.shape
````

````python
parda = enem_campinas.loc[enem_campinas.RACA == 'parda'].sample(frac=0.15)
parda.shape
````

````python
preta = enem_campinas.loc[enem_campinas.RACA == 'preta'].sample(frac=0.15)
preta.shape
````

````python
amarela = enem_campinas.loc[enem_campinas.RACA == 'amarela'].sample(frac=0.15)
amarela.shape
````

````python
nd = enem_campinas.loc[enem_campinas.RACA == 'nao_declarado'].sample(frac=0.15)
nd.shape
````

````python
indigena = enem_campinas.loc[enem_campinas.RACA == 'indígena'].sample(frac=0.15)
indigena.shape
````

````python
amos_estra = pd.concat([branca, parda, preta, amarela, nd, indigena]) # concatenando as variáveis
````

````python
amos_estra
````

**Grandes quantidades de grupos**
Para grandes grupos, fica inviável na maneira acima. Fazemos o seguinte:
