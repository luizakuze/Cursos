# Amostragem Aleatória Simples
Seleção por sorteio, sem filtros.

````python
np.random.seed(1) # gera valor aleatório e sempre que rodar vai dar o mesmo número
````

- np.random.seed é usado para inicializar o gerador de números aleatórios do NumPy com uma semente específica.
- Serve para reproduzir resultados aleatórios em diferentes execuções do código.
- A semente deve ser definida antes de qualquer chamada a funções que geram números aleatórios.
- Se definida corretamente, produzirá a mesma sequência de números pseudoaleatórios em diferentes execuções.
- Se definida após a geração de números aleatórios, não terá efeito retroativo na sequência já gerada.
- Se, mesmo após definir a semente, os resultados continuarem diferentes, revise o código para garantir que todas as fontes de aleatoriedade sejam tratadas adequadamente.


````python
# pega uma fração aleatória de enem campinas (amostragem aleatória) e pega 20% dos dados
amos_simples = enem_campinas.sample(frac=0.20) # porcentagem
````

# Amostragem Sistemática
Membros da população ordenados numericamente e selecionados de acordo com uma sequência numérica.

````python
enem_campinas.shape
````

````python
np.random.seed(2)
inicio = np.random.choice(10,1) # (limite do número, quantidade de números) seleciona número inicial p pegar elementos
````

````python
inicio
````

````python
amostra2 = np.arange(inicio,13198,100) # (valor numérico inicial, valor total de linhas, sequência de x números em x números)
amostra2
````

````python
# daqui para cima tinhamos somente os números...

# queremos isso tabelado 

# "pegue na tabela enem_campinas os números da amostra2"
# iloc busca pelo índice de amostra2 na tabela
# ,: para todas as colunas referentes aos registros/índices
amos_sistematica = enem_campinas.iloc[amostra2,:] # 
````

````python
amos_sistematica
````
