# Distribui√ß√£o de Frequ√™ncias üìà

```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
enem_sp = pd.read_csv('enem_2019_tratado.csv', sep=',', encoding='iso-8859-1')
```

```python
enem_sp.head()
```

```python
enem_sorocaba = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Sorocaba']
```

```python
enem_sorocaba.head()
```

```python
enem_sorocaba.shape
```

```python
idades = enem_sorocaba['IDADE']
idades
```

```python
freq_absoluta = Counter(idades)
freq_absoluta # est√° em formato de dicion√°rio
```

```python
# transformando o dicion√°rio em dataframe, sendo ele orientado por um √≠ndice
freq_absoluta = pd.DataFrame.from_dict(freq_absoluta, orient='index')
freq_absoluta # aqui est√° um pouco desorganizado...
```

```python
freq_absoluta = freq_absoluta.sort_index(ascending=True)
freq_absoluta # ordenando em ordem crescente
```

```python
freq_relativa = freq_absoluta / freq_absoluta.sum()
freq_relativa # obtendo a frequ√™ncia relativa
```

```python
freq_relativa_perc = freq_relativa * 100
freq_relativa_perc # obtendo frequ√™ncia percentual relativa
```

# Histograma üìä

## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
enem_sorocaba.IDADE.hist(bins=8) #tabela.COLUNA.grafico(bins=num colunas)
plt.title('Distribui√ß√£o de Idades - Enem SP 2019 - Sorocaba')
plt.show() # um histograma: eixo Y contagem e eixo X classe
```

```python
plt.hist(enem_sorocaba.IDADE, 8, rwidth=1) # plt.hist(tabela.COLUNA, num faixas/colunas, espa√ßamento entre colunas)
plt.title('Distribui√ß√£o de Idades - Enem SP 2019 - Sorocaba')
plt.show()
```

```python
from matplotlib import colors
# eixo Y √© frequ√™ncia e eixo X a classe
plt.hist(enem_sorocaba.IDADE, 10, rwidth=1, color='pink', density = True) # (tabela.COLUNA, bins (n colunas), separa√ß√£o, cor, frequ√™ncia relativa)
plt.title('Distribui√ß√£o de Idades - Enem SP 2019 - Sorocaba')
plt.xlabel ('Idade') # r√≥tulo X
plt.ylabel ('Frequ√™ncia') # r√≥tulo Y
plt.show()
```
