```python
# criando uma função
def probab (A, E):
    resultado = (A/E) * 100 # (casos favoráveis(n elementos)/casos totais(espaço amostral)) * 100
    print('{:.2f}%'.format(resultado))
```

```python
# probabilidade de retirar um vestibulando de Guarulhos
probab (len(enem_guarulhos), len(enem_sp))
```

```python
# probabilidade de retirar um vestibulando de Bauru
probab (len(enem_bauru), len(enem_sp))
```

## Probabilidade d enão ocorrer um evento

```python
# função probabilidade de NÃO ocorrer um evento
def probab_nao (A, E):
  resultado = (1- (A / E))*100 
  print('{:.2f}%'.format(resultado))
```

```python
# probabilidade de NÃO retirar um vestibulando de Guarulhos
probab_nao (len(enem_guarulhos), len(enem_sp))
```

```python
# probabilidade de NÃO retirar um vestibulando de Bauru
probab_nao (len(enem_bauru), len(enem_sp))
```

## Probabilidade da união de eventos mutuamente exclusivos (AUB)

Ocorre A  OU  B.
Elementos que não ocorrem simultâneamente.

```python
# função probabilidade da ocorrência de A ou B
def probab_uniao (A, B, E):
    resultado = (A/E + B/E)*100 
    print('{:.2f}%'.format(resultado))
```

```python
# probabilidade de retirar um vestibulando de Guarulhos OU Bauru
probab_uniao(len(enem_guarulhos),len(enem_bauru),len(enem_sp))
```

## Probabilidade Condicional

Probabilidade de ocorrer A dado que B ocorreu.

```python
# função para calcular a probabilidade condicional P(A|B)
def probab_cond (AB, B):
    resultado = (AB / B)*100 # AB é A intersecção com B
    print('{:.2f}%'.format(resultado))
```

```python
# procurando a probabilidade de ser raça PARDA (A) dado que foi retirada uma MULHER (B)

# extraindo a mulher
mulher = enem_sp.loc[enem_sp.SEXO == 'F']
```

```python
# extraindo mulher parda
mulher_parda = mulher.loc[mulher.RACA == 'parda']
```

```python
mulher_parda.head(20)
```

```python
probab_cond(len(mulher_parda),len(mulher))
```
