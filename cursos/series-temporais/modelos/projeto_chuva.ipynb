````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 16, 5
````

````python
chuva = pd.read_csv('chuva_tratado.csv', sep=',')
chuva.head()
````

## Cria√ß√£o da S√©rie

````python
# excluindo a coluna ano
chuva2 = chuva.drop(columns='Ano')
chuva2.head()
````

````python
# como colocar os valores referentes ao primeiro ano? e formatar com a sequ√™ncia de Janeiro, Fevereiro, Mar√ßo,..., Dezembro

# transformar em array??
chuva3 = chuva2.values
chuva3 # forma v√°rios vetores dentro do array, n√£o queremos essa quebra. 
````

````python
# para evitar essa quebra, transformamos em uma lista
chuva4 = list(chuva3.flatten()) # sem quebra, ent√£o utilizamos o m√©todo flatten
print(chuva4) # ficou tudo em somente uma linha
````

````python
# montando os √≠ndices mensalmente
indice = pd.date_range('1985', periods = len(chuva4), freq = 'M')
indice
````

````python
serie = pd.Series(chuva4, index = indice)
serie
````

````python
serie.plot()
plt.show()
````

## An√°lise e Prepara√ß√£o da S√©rie

````python
### **M√©dia M√≥vel**
````

````python
media_movel = serie.rolling(window=6)
media_movel = media_movel.mean()
````

````python
plt.plot(media_movel);
````

````python
plt.plot(serie, label='S√©rie Real')
plt.plot(media_movel,color='red', label='M√©dia M√≥vel 6 meses')
plt.legend(loc='best')
plt.show() # n√£o h√° problemas de tend√™ncia
````

### **Normalidade e Transforma√ß√£o**

````python
import scipy.stats as stats
````

````python
stats.probplot(serie, dist="norm", plot=plt)
plt.title("Normal QQ plot")
plt.show()
````

**Teste Shapiro-Wilk**

Crit√©rios:

- n√≠vel de signific√¢ncia de 0,05

- Ho = distribui√ß√£o normal p > 0,05

- Ha = distribui√ß√£o n√£o normal p <= 0,05

````python
e, p = stats.shapiro(serie)
print(f'Estat√≠stica de teste: {e}')
print(f'p-valor: {p}')
````

````python
# transforma√ß√£o por log (Diminuir vari√¢ncia e melhorar normalidade)
serie2 = np.log(serie)
serie2
````

````python
e, p = stats.shapiro(serie2)
print(f'Estat√≠stica de teste: {e}') # n√£o pode ter valor 0 e nem valores negativos, essa transforma√ß√£o n√£o √© poss√≠vel nesse caso
print(f'p-valor: {p}')
````

````python
# transforma√ß√£o por raiz c√∫bica (quando possui dados com valor zero ou negativos)
serie3 = np.sign(serie)*abs(serie)**(1/3)
serie3
````

````python
stats.probplot(serie3, dist="norm", plot=plt)
plt.title("Normal QQ plot")
plt.show()
````

````python
e, p = stats.shapiro(serie3)
print(f'Estat√≠stica de teste: {e}')
print(f'p-valor: {p}')
````

### **Estacionaridade**

(m√©dia constante)

````python
import statsmodels.tsa.stattools
````

Teste KPSS (Kwiatkowski-Phillips-Schmidt-Shin)

Ho = n√£o √© estacion√°rio: estat√≠stica do teste > valor cr√≠tico

Ha = √© estacion√°rio:  estat√≠stica do teste < valor cr√≠tico

````python
kpss = statsmodels.tsa.stattools.kpss(serie3)
print('Estat√≠tica do teste: {:.4f}'.format(kpss[0]))
print('p_valor: {:.4f}'.format(kpss[1]))
print('Valores Cr√≠ticos:')
for chave, valor in kpss[3].items():
   print('{}: {:.4f}'.format(chave, valor))
````

### **Autocorrela√ß√£o**

````python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
````

````python
plot_acf(serie3, lags=20)
plt.show() # h√° autocorrela√ß√£o, tem um padr√£o no desenho gr√°fico! forte sazonalidade.
````

````python
plot_pacf(serie, lags=20) # verificando quando est√£o intercalados.. ainda registros autocorrelacionados, fora do intervalo de confian√ßa.
plt.show() 
````
