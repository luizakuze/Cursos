# Criação de um passeio aleatório           

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
from random import sample, random
```

```python
# recebe uma amostragem de um total de 100 seleciona aleatoriamente 41 valores
dados1 = sample(range(100), k=41)
dados1
```

```python
# agora pegando de uma extensão de 1980 até 2021 
anos = list(range(1980, 2021))
anos
```

```python
# criando a série, agora não é a partir de tabelas como já fizemos
serie1 = pd.Series(dados1, index = anos)
serie1
```

```python
type(serie1)
```

```python
serie1.plot() # poodemos analisar que não há um padrão! série temporal imprevisível!
plt.show()
```

```python
# série passeio aleatório

## criando de uma forma diferente
## "mais formal"

# criando uma lista
serie2 = list()

# incluindo valor aleatóri: -1 se o valor aleatório for < 0.5 se não, incluir 1
serie2.append(-1 if random() < 0.5 else 1)

# fazendo isso para uma maior quantidade de valores
for i in range (1, 1000):
    movimento = -1 if random() < 0.5 else 1
    
    valor = serie2[i-1] + movimento
    serie2.append(valor)
```

```python
random() # visualizando aleatoriedade
```

```python
serie2 = pd.Series(serie2) # transformando em série
serie2.plot()
plt.show()
```
