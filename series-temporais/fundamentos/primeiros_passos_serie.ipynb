# Série Anual

````python
# série de 1980 a 2020 (41 anos, contando os próprios limites)
np.random.seed(10) # criando uma semente para sempre manter esses mesmos dados 
dados1 = np.random.normal(0,1,41) # criando variável aleatória com distribuição normal, parâmetros: (média aproximada, desvio padrão, quantos valores)
dados1
````

````python
type(dados1) # tipo de dados1
````

````python
serie = pd.Series(dados1) # transformando em uma série
````

````python
type(serie)
````

````python
serie
````

````python
serie.plot()
plt.show() # mostrando a série
````

não ficou muito legal o plot acima, já que no eixo x não está mostrando o ano a cada medição. vamos passar para um dataframe

## Diferença de série e dataframe

### Série
- uma variável no conjunto dados.
- vista como uma coluna de uma tabela.
- ```data = [1, 2, 3, 4, 5]```

### Dataframe
- várias séries.
- vista como planilha, várias colunas.
- ```data = {'Coluna1': [1, 2, 3, 4, 5], 'Coluna2': ['a', 'b', 'c', 'd', 'e']}```

````python
dados1 = pd.DataFrame(dados1)
dados1 # dados1 em formato de tabela
````

````python
dados1.columns = ['valores']
dados1.head()
````

````python
dados1.shape # quantas linhas e colunas
````

````python
dados1.describe # média, desevio padrão, mínimo, máximo, ...
````

````python
# criando o índice das datas

# parâmetros = (onde inicia, onde termina, frequência anual/mensal/semanal)
# periods = len(dados1) para facilitar, deixar o próprio python verificar o limite máximo
indice  = pd.date_range('1980', periods = len(dados1), freq = 'Y') 
indice
````

````python
serie1 = pd.Series(dados1['valores'].values, index = indice) # criando uma série, parâmetros = (valores da coluna, índices x)
serie1 
````

````python
type(serie1)
````

````python
serie1.plot()
plt.show()
````

````python
import scipy.stats as stats # parte de estatística
````

````python
# verificando se é uma distribuição normal
stats.probplot(serie1, dist='norm', plot=plt)
plt.title("Normal QQ Plot")
plt.show()
````

Verificando a partir de valores numéricos para confirmar normalidade

Critérios:

- Nível de significância de 0,05;

- Quando p > 0,05 é uma distribuição normal.

````python
e, p = stats.shapiro(serie1) # e = estatística do teste, p = p-valor
print(f'Estatística de teste: {e}')
print(f'p-valor: {p}') # é uma distribuição normal! p > 0,05
````

# Série Mensal

````python
np.random.seed(6)
dados2 = np.random.normal(0,1,72) # lembrando que isso me dá um array
dados2
````

````python
type(dados2)
````

````python
dados2 = pd.DataFrame(dados2)
dados2.head()
````

````python
dados2.columns = ['valores']
dados2.head()
````

````python
dados2.describe()
````

````python
dados2.shape
````

````python
import datetime # criando o período
````

````python
data = np.array('2015-01', dtype = np.datetime64()) # (início, valor final ??)
data
````

````python
data = data + np.arange(72) # sobreescrevendo data com a data inicial e com amis 72 valores de data
data
````

````python
data = pd.DataFrame(data)
data
````

````python
data.columns = ['data']
data.head()
````

````python
data.shape
````

````python
serie2 = pd.concat([data, dados2], axis=1) # axis = 1 são clunas, axis = 0 são linhas
serie2 # ainda é um dataframe!!
````

````python
# transformando efetivamente em uma série
# tendo como valores a coluna 'valores' e como índices a coluna 'data'
serie2 = pd.Series(serie2['valores'].values, index = serie2['data'])
serie2
````

````python
type(serie2)
````

````python
serie2.plot()
plt.show()
````

Analisando a normalidade

````python
stats.probplot(serie2, dist='norm', plot=plt)
plt.title('Normal QQ plot')
plt.show()
````

Teste Shapiro-Wilk

Critérios:

- Nível de significância de 0,05 
- Quando p > 0,05 é distribuição normal

````python
e, p = stats.shapiro(serie2)
print(f'Estatística de teste: {e}')
print(f'p-valor: {p}')
````

## Série diária
